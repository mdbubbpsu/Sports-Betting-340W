```python
import pandas as pd
import numpy as np
from pandas import Timestamp
```

```python
games = pd.read_csv(
        '/Users/mattbubb/Spring_2021/DS_340W/Sports-Betting-340W/Data/nba_games_all.csv')
# Sort immediately 
games = games.sort_values('game_date').dropna().reset_index().drop('index', axis = 1)
games
```

```python
# Grab date
games[['Year', 'Month', 'Day']] = games['game_date'].str.split('-', 2, expand = True)
games
```

```python
teams = pd.read_csv(
        '/Users/mattbubb/Spring_2021/DS_340W/Sports-Betting-340W/Data/nba_teams_all.csv')
teams = teams.dropna().drop(['league_id','min_year', 'max_year'],axis =1)
```

```python
# Merge with Team data to get abbreviations of teams
games = games.merge(teams, on = 'team_id')
games
```

```python
# Group by Teams, Seasons, and Month/date to get a cumulative sum of stats to be divided by number of games later
#needsDivision = games.groupby(['abbreviation', 'season', 'Month','Day']).sum().groupby(['abbreviation','season']).cumsum().reset_index()
#needsDivision

needsDivision = games.groupby(['abbreviation', 'season', 'game_date']).sum().groupby(['abbreviation','season']).cumsum().reset_index()
needsDivision
```

```python
# Loop through rows to add a column which adds 1 for each game date to divide the cumulative stats by # of games
test = games.groupby(['abbreviation', 'season','game_date']).size().reset_index(name='numGamesPerSeason')
test = test.sort_values(['abbreviation','season'])
test['game_date'] = pd.to_datetime(test['game_date'])
first = ('ATL', '2003-04')
firstGame = Timestamp('2003-11-01 00:00:00')
test['gameNo'] = 0
lst = []
count = 1
for group_name, df_group in test.groupby(['abbreviation', 'season']):
    for index, row in df_group.iterrows():
        # If not first in group
        if group_name != first:
            firstGame = row['game_date']
            count = 1
            lst.append(count)
            count += 1
            first = group_name
        else:
            lst.append(count)
            count += 1

test['gameNo'] = lst
```

```python
# Game number per team
test.drop(['season', 'numGamesPerSeason'], axis = 1, inplace=True)
test.head(83)
```

```python
# Sort Values
games = games.sort_values(['abbreviation', 'game_date'])
games.reset_index().drop('index', axis = 1, inplace = True)
games
```

```python
# Merge number of games with Games Df
games['game_date'] = pd.to_datetime(games['game_date'])
newDf = games.merge(test, on = ['abbreviation', 'game_date'])
newDf
```

```python
# Pull out game number columns
gameNumber = newDf['gameNo']
gameNumber
```

```python
# Divide number of cumulative stats by number of games at any given time
# Add back in some important qualitative columns
needsDivision['gameNumber'] = gameNumber
upToDateStats = needsDivision[['w',
       'l', 'w_pct', 'min', 'fgm', 'fga', 'fg_pct', 'fg3m', 'fg3a', 'fg3_pct',
       'ftm', 'fta', 'ft_pct', 'oreb', 'dreb', 'reb', 'ast', 'stl', 'blk',
       'tov', 'pf', 'pts']].div(needsDivision.gameNumber, axis = 0).where(needsDivision.gameNumber != 0)
upToDateStats['abbreviation'] = newDf['abbreviation']
upToDateStats['season']=newDf['season']
upToDateStats['game_id'] = newDf['game_id']
upToDateStats['team_id'] = newDf['team_id']
upToDateStats['gameNo'] = needsDivision['gameNumber']
upToDateStats['game_date'] = needsDivision['game_date']
upToDateStats['is_home'] = newDf['is_home']
```

```python
upToDateStats = upToDateStats
#upToDateStats.drop('index', axis = 1, inplace = True)
upToDateStats.head(83)
```

```python
upToDateStats['ShiftedGameID'] = upToDateStats.groupby(['abbreviation', 'season'])['game_id'].shift(-1)

upToDateStats['ShiftedGameNumber'] = upToDateStats.groupby(['abbreviation', 'season'])['gameNo'].shift(-1)
upToDateStats = upToDateStats.dropna()
```

```python
upToDateStats.loc[:,'ShiftedGameNumber'] = upToDateStats.ShiftedGameNumber.apply(int)
upToDateStats.loc[:,'ShiftedGameID'] = upToDateStats.ShiftedGameID.apply(int)
upToDateStats.head(83)
```

```python
upToDateStats.drop(['game_id', 'gameNo'], axis = 1, inplace = True)
```

```python
upToDateStats = upToDateStats [['abbreviation','team_id','is_home','season','game_date','ShiftedGameNumber','ShiftedGameID','w', 'l', 'w_pct', 'min', 'fgm', 'fga', 'fg_pct', 'fg3m', 'fg3a',
       'fg3_pct', 'ftm', 'fta', 'ft_pct', 'oreb', 'dreb', 'reb', 'ast', 'stl',
       'blk', 'tov', 'pf', 'pts' 
          ]]
```
